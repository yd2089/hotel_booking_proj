## Analysis of booking and cancellation patterns

```python
df['children'].fillna(0, inplace=True)

hotel_cancellation = df.groupby('hotel')['is_canceled'].mean().reset_index()
hotel_cancellation['cancel_percentage'] = hotel_cancellation['is_canceled'] * 100

plt.figure(figsize=(10, 6))
sns.barplot(x='hotel', y='cancel_percentage', data=hotel_cancellation)
plt.title('Cancellation Rate by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Cancellation Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
print(hotel_cancellation)
```

```python
monthly_bookings = df.groupby(['arrival_date_month', 'hotel']).size().reset_index(name='bookings')
monthly_cancellations = df.groupby(['arrival_date_month', 'hotel']).agg({'is_canceled': 'sum'}).reset_index()
monthly_cancellations['cancellation_rate'] = monthly_cancellations['is_canceled'] / monthly_bookings['bookings'] * 100

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_bookings['arrival_date_month'] = pd.Categorical(monthly_bookings['arrival_date_month'], categories=month_order, ordered=True)
monthly_cancellations['arrival_date_month'] = pd.Categorical(monthly_cancellations['arrival_date_month'], categories=month_order, ordered=True)

monthly_bookings = monthly_bookings.sort_values('arrival_date_month')
monthly_cancellations = monthly_cancellations.sort_values('arrival_date_month')

plt.figure(figsize=(14, 7))
sns.lineplot(x='arrival_date_month', y='cancellation_rate', hue='hotel', data=monthly_cancellations, marker='o')
plt.title('Cancellation Rate by Month and Hotel Type')
plt.xlabel('Month')
plt.ylabel('Cancellation Rate (%)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
result_df = pd.merge(monthly_bookings, monthly_cancellations, on=['arrival_date_month', 'hotel'])
print(result_df[['arrival_date_month', 'hotel', 'bookings', 'is_canceled', 'cancellation_rate']])
```

## Customer profile analysis

## Booking Characteristics Analysis

```python
df['lead_time_group'] = pd.cut(df['lead_time'], 
                              bins=[0, 7, 30, 90, 180, 365, 800], 
                              labels=['0-7 days', '8-30 days', '31-90 days', '91-180 days', '181-365 days', '366+ days'])

lead_time_analysis = df.groupby('lead_time_group').agg(
    total_bookings=('is_canceled', 'count'),
    cancellations=('is_canceled', 'sum'),
).reset_index()

lead_time_analysis['cancellation_rate'] = lead_time_analysis['cancellations'] / lead_time_analysis['total_bookings'] * 100

plt.figure(figsize=(12, 6))
sns.barplot(x='lead_time_group', y='cancellation_rate', data=lead_time_analysis)
plt.title('Cancellation Rate by Lead Time Group')
plt.xlabel('Lead Time')
plt.ylabel('Cancellation Rate (%)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
print(lead_time_analysis)
```

```python
market_segment_analysis = df.groupby('market_segment').agg(
    bookings=('is_canceled', 'count'),
    cancellations=('is_canceled', 'sum')
).reset_index()

market_segment_analysis['cancellation_rate'] = market_segment_analysis['cancellations'] / market_segment_analysis['bookings'] * 100
market_segment_analysis = market_segment_analysis.sort_values('bookings', ascending=False)

channel_analysis = df.groupby('distribution_channel').agg(
    bookings=('is_canceled', 'count'),
    cancellations=('is_canceled', 'sum')
).reset_index()

channel_analysis['cancellation_rate'] = channel_analysis['cancellations'] / channel_analysis['bookings'] * 100
channel_analysis = channel_analysis.sort_values('bookings', ascending=False)

plt.figure(figsize=(12, 8))

ax1 = plt.subplot(2, 1, 1)
sns.barplot(x='market_segment', y='bookings', data=market_segment_analysis, ax=ax1)
ax1.set_title('Number of Bookings by Market Segment')
ax1.set_xlabel('')
ax1.set_ylabel('Number of Bookings')
ax1.tick_params(axis='x', rotation=45)

for i, row in enumerate(market_segment_analysis.itertuples()):
    ax1.text(i, row.bookings + 1000, f'{row.cancellation_rate:.1f}%', 
             ha='center', va='bottom', color='black', fontweight='bold')

ax2 = plt.subplot(2, 1, 2)
sns.barplot(x='distribution_channel', y='bookings', data=channel_analysis, ax=ax2)
ax2.set_title('Number of Bookings by Distribution Channel')
ax2.set_xlabel('Distribution Channel')
ax2.set_ylabel('Number of Bookings')
ax2.tick_params(axis='x', rotation=45)

for i, row in enumerate(channel_analysis.itertuples()):
    ax2.text(i, row.bookings + 1000, f'{row.cancellation_rate:.1f}%', 
             ha='center', va='bottom', color='black', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
status_mapping = {
    'Canceled': 1,
    'No-Show': 1,  # Assuming No-Show should be considered cancellations
    'Check-Out': 0
}

# Create corrected cancellation indicator
df['is_cancelled_corrected'] = df['reservation_status'].map(status_mapping)

# Handle any unknown statuses (fill with original is_canceled as fallback)
df['is_cancelled_corrected'] = df['is_cancelled_corrected'].fillna(df['is_canceled'])
```

```python
# 2. Analyze cancellation rates by deposit type
deposit_analysis = df.groupby('deposit_type').agg(
    total_bookings=('is_cancelled_corrected', 'size'),
    cancellations=('is_cancelled_corrected', 'sum'),
    status_distribution=('reservation_status', lambda x: x.value_counts().to_dict())
).reset_index()

deposit_analysis['cancellation_rate'] = (deposit_analysis['cancellations'] / deposit_analysis['total_bookings']) * 100
```

```python
# 3. Create visualization with status breakdown
plt.figure(figsize=(14, 8))

# Bar positions and width
x = range(len(deposit_analysis))
width = 0.8

# Plot cancellation rates
bars = plt.bar(x, deposit_analysis['cancellation_rate'], width, color='#43739d', edgecolor='black')

# Add status distribution annotations
for i, row in deposit_analysis.iterrows():
    dist_text = "\n".join([f"{k}: {v}" for k, v in row['status_distribution'].items()])
    plt.text(x=i, y=row['cancellation_rate'] + 2, 
             s=dist_text, ha='center', va='bottom', fontsize=9)

plt.title('Validated Cancellation Rates by Deposit Type\n(with Reservation Status Breakdown)', fontsize=14, pad=20)
plt.xlabel('Deposit Type', fontsize=12)
plt.ylabel('Cancellation Rate (%)', fontsize=12)
plt.xticks(x, deposit_analysis['deposit_type'], rotation=45, ha='right', fontsize=11)
plt.ylim(0, 110)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add percentage labels
for bar, rate in zip(bars, deposit_analysis['cancellation_rate']):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height-5,
             f'{rate:.1f}%', ha='center', va='top', color='white', fontweight='bold')

plt.tight_layout()
plt.show()

# Print validation summary
print("Data Validation Summary:")
print("========================")
print(f"Original vs Corrected mismatches: {(df['is_canceled'] != df['is_cancelled_corrected']).sum()} records")
print("\nStatus Distribution:")
print(df['reservation_status'].value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Selecting the relevant features
selected_features = [
    "lead_time",
    "stays_in_weekend_nights",
    "stays_in_week_nights",
    "adults",
    "children",
    "previous_cancellations",
    "previous_bookings_not_canceled",
    "days_in_waiting_list",
    "adr",
]

# Extract relevant columns
df_cluster = df[selected_features].copy()

# Handle missing values by filling with median
df_cluster = df_cluster.fillna(df_cluster.median())

# Normalize the data
scaler = StandardScaler()
df_cluster_scaled = scaler.fit_transform(df_cluster)

# Determine optimal number of clusters using Elbow method
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_cluster_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve to determine the optimal number of clusters
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker="o", linestyle="-")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal K")
plt.grid(True)
plt.show()
```

There is no clear elbow we can choose. In this case, we choose only `adr` and `lead_time` to do clustering.

```python
# Selecting only 'lead_time' and 'adr' for clustering
df_cluster_2d = df[["lead_time", "adr"]].copy()

# Handle missing values by filling with median
df_cluster_2d = df_cluster_2d.fillna(df_cluster_2d.median())

# Normalize the data
scaler = StandardScaler()
df_cluster_2d_scaled = scaler.fit_transform(df_cluster_2d)

# Determine optimal number of clusters using Elbow method
inertia_2d = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_cluster_2d_scaled)
    inertia_2d.append(kmeans.inertia_)

# Plot the Elbow curve for 'lead_time' and 'adr' clustering
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia_2d, marker="o", linestyle="-")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal K (lead_time & adr)")
plt.grid(True)
plt.show()
```

```python
# Apply K-Means clustering with K=3
kmeans_2d = KMeans(n_clusters=3, random_state=42, n_init=10)
df_cluster_2d["cluster"] = kmeans_2d.fit_predict(df_cluster_2d_scaled)
# Get cluster centroids (inverse transform to original scale)
centroids = scaler.inverse_transform(kmeans_2d.cluster_centers_)

# Create the scatter plot with labeled clusters
plt.figure(figsize=(8, 6))

# Plot each cluster separately
for cluster in range(3):
    cluster_data = df_cluster_2d[df_cluster_2d["cluster"] == cluster]
    plt.scatter(cluster_data["lead_time"], cluster_data["adr"], label=f"Cluster {cluster}", alpha=0.6)

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c="black", marker="X", s=200, label="Centroids")

# Labels and title
plt.xlabel("Lead Time")
plt.ylabel("ADR (Average Daily Rate)")
plt.title("K-Means Clustering (k=3)")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
